```python
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Override pandas datareader method
yf.pdr_override()

# Define the tickers  
# load the data: Apple, Google, Microsoft, Amazon
tickers = ['AAPL', 'GOOG', 'MSFT', 'AMZN']

# Define the start and end dates
start_date = '2022-01-01'
end_date = '2022-12-31'

# Download the data
stock_data = {ticker: pdr.get_data_yahoo(ticker, start=start_date, end=end_date
                                        ) for ticker in tickers}
```

### Returns

```python
# Create an empty DataFrame
returns = pd.DataFrame()

# Calculate percent change for each stock and add it to the DataFrame
for ticker, data in stock_data.items():
    returns[ticker] = data['Close'].pct_change()

# Display the DataFrame
print(returns)
```

## Closing Price


```python
import matplotlib.pyplot as plt

# Plot the closing prices for each ticker
for ticker, data in stock_data.items():
    plt.plot(data.index, data['Close'], label=ticker)

# Customize the plot
plt.title('Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Display the plot
plt.show()
```

### Volume of Sales


```python
# Now let's plot the total volume of stock being traded each day
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, (ticker, data) in enumerate(stock_data.items()):
    plt.subplot(2, 2, i+1)
    data['Volume'].plot()
    plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title(f"Closing Price of {ticker}", fontsize=16)  # Increase the font size of the title
    plt.grid(True)
    
plt.tight_layout()
```

### Moving Average 

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

tickers = ['AAPL', 'GOOG', 'MSFT', 'AMZN']
ma_days = [10, 20]  # Moving average days

for i, ticker in enumerate(tickers):
    data = stock_data[ticker]
    axs[i//2, i%2].plot(data.index, data['Adj Close'], label='Adj Close')
    
    for ma in ma_days:
        column_name = f"MA for {ma} days"
        data[column_name] = data['Adj Close'].rolling(ma).mean()
        axs[i//2, i%2].plot(data.index, data[column_name], label=column_name)

    axs[i//2, i%2].set_title(ticker)
    axs[i//2, i%2].legend()

plt.tight_layout()
plt.show()
```

### Daily Stock Return

```python
returns_data = {}

for ticker, data in stock_data.items():
    returns_data[ticker] = data['Adj Close'].pct_change()

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

for i, (ticker, data) in enumerate(returns_data.items()):
    axes[i//2, i%2].plot(data.index, data, label='Daily Return')
    axes[i//2, i%2].set_title(ticker)
    axes[i//2, i%2].legend()

plt.tight_layout()
plt.show()
```

### Average Daily Return using Histogram  

```python
plt.figure(figsize=(12, 9))

for i, (ticker, data) in enumerate(returns_data.items()):
    plt.subplot(2, 2, i+1)
    data.hist(bins=50)
    plt.ylabel('Counts')
    plt.xlabel('Daily Return')
    plt.title(f"{ticker}", fontsize=16)
    plt.grid(True)
    
plt.tight_layout()
plt.show()
```

### Volatility

```python
# Compute and store the logarithmic returns for each stock
log_return = {ticker: np.log(stock_data[ticker]['Close'] / stock_data[ticker]['Close'].shift(1)) for ticker in tickers}

# Compute and store the volatility for each stock
volatility = {ticker: log_return[ticker].std() * np.sqrt(252) for ticker in tickers}

# Print the volatility for each stock
for ticker in tickers:
    print(f"Volatility for {ticker}: {volatility[ticker]}")
```

### Comparative Returns

```python
# Compute and store the daily returns for each stock
daily_return = {ticker: stock_data[ticker]['Close'].pct_change() for ticker in tickers}

# Compute and store the cumulative returns for each stock
cumulative_return = {ticker: (1 + daily_return[ticker]).cumprod() for ticker in tickers}

# Plot the cumulative returns
plt.figure(figsize=(12, 8))

for ticker in tickers:
    plt.plot(cumulative_return[ticker], label=ticker)

plt.legend(loc='best')
plt.xlabel('Date')
plt.ylabel('Growth of $1 investment')
plt.title('Cumulative Returns of Tech Stocks Over Time')
plt.show()
```

### Correlation 

```python
# Calculate the daily returns
daily_returns = pd.DataFrame({ticker: stock_data[ticker]['Adj Close'].pct_change() for ticker in tickers})

# Calculate the correlation matrix
corr_matrix = daily_returns.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
